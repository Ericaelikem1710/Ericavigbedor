```python
alcohol=head["alcohol"]
quality=head["quality"]
fixedAcidity=head["fixed acidity"]
volatileAcidity=head['volatile acidity']
pH=head["pH"]
```

### Results And Discussion

```python
plt.hist(alcohol)
plt.xlabel("Alcohol Content")
plt.ylabel("Frequency")
plt.title("A Graph Of Alcohol Against Frequency")
```

```python
plt.hist(quality)
plt.xlabel("Wine Quality")
plt.ylabel("Frequency")
plt.title("A Graph Of Quality Against Frequency")
```

```python
plt.hist(fixedAcidity)
plt.xlabel("Fixed Acidity")
plt.ylabel("Frequency")
plt.title("A Graph Of Fixed Acidity Against Frequency")
```

```python
plt.hist(volatileAcidity)
plt.xlabel("Volatile Acidity")
plt.ylabel("Frequency")
plt.title("A Graph Of Volatile Acidity Against Frequency")
```

```python
plt.hist(pH)
plt.xlabel("pH Content")
plt.ylabel("Frequency")
plt.title("A Graph Of pH Against Frequency")
```

```python
plt.scatter(alcohol,quality)
plt.xlabel("Alcohol Content")
plt.ylabel("quality")
plt.title("A Graph Of Alcohol Content Against Quality")
```

```python
plt.scatter(fixedAcidity,quality)
plt.xlabel("Fixed Acidity")
plt.ylabel("quality")
plt.title("A Graph Of Fixed Acidity Against Quality")
```


### Histogram
The histogram displays the distribution of alcohol content in the wine dataset. <b>The x-axis</b> represents the alcohol content, divided into 20 bins. The y-axis shows the frequency of occurrences within each bin. The distribution appears somewhat right-skewed, indicating a higher frequency of wines with lower alcohol content.

### Scatter plot 
The scatter plot explores the relationship between alcohol content and wine quality. Each point represents a wine, with alcohol content on the x-axis and quality on the y-axis. There seems to be a positive correlation, suggesting that wines with higher alcohol content might be associated with higher quality. However, the spread of points indicates that other factors also contribute to wine quality.
To sum up, these visualizations provide a thorough knowledge of the dataset's alcohol content, assisting in the comprehension of its distribution, central tendency, variability, and any correlations with other variables.
